```python
import pandas as pd
import re

#Read Data in

file_path = os.path.expanduser('~/Downloads/BusinessAnalyst.xlsx')


df = pd.read_excel(file_path)

#Clean salary and get an estimated average column


def remove_glassdoor_est(salary_str):
    return salary_str.replace(' (Glassdoor est.)', '')


df['Cleaned Salary Estimate'] = df['Salary Estimate'].apply(remove_glassdoor_est)


def process_salary(salary_str):
 
    salary_str = salary_str.replace('$', '').replace('k', '').replace('K', '').replace(' ', '')
    
  
    salary_range = salary_str.split('-')
    

    try:
        if len(salary_range) == 2:
            low_salary = int(re.sub(r'\D', '', salary_range[0]))
            high_salary = int(re.sub(r'\D', '', salary_range[1]))
           
            average_salary = (low_salary + high_salary) / 2
        elif len(salary_range) == 1:
            average_salary = int(re.sub(r'\D', '', salary_range[0]))
        else:
            return None
    except ValueError:
        return None
    
    return average_salary


df['Estimated Average Salary'] = df['Cleaned Salary Estimate'].apply(process_salary)



df = df.dropna(subset=['Estimated Average Salary'])


print(df[['Salary Estimate', 'Cleaned Salary Estimate', 'Estimated Average Salary']].head())
```

```python
#Data cleaning for company name

def clean_company_name(company_name):
    
    company_name = str(company_name)
   
    return re.sub(r'\n\d+.*', '', company_name)


df['Cleaned Company Name'] = df['Company Name'].apply(clean_company_name)
df
```

```python
# Replace 1 with "Yes" and -1 with "No" in the "Easy Apply" column
df['Easy Apply'] = df['Easy Apply'].replace({1: 'Yes', -1: 'No'})

print(df[['Easy Apply']].head())
```

```python
import numpy as np

# Replace all -1 values in the entire DataFrame with NaN
df.replace(-1, np.nan, inplace=True)

# Display the first few rows of the DataFrame to verify the results
print(df.head())
```

```python
# We want job title, Estimated Average Salary, Job Description top words, rating, Location, Size, 
new_df = df[['Estimated Average Salary', 'Rating', 'Cleaned Company Name', 'Location', 'Size', 'Sector', 'Easy Apply']]

#Verify data is ready to be visualized and manipulated
#new_df['Rating'].value_counts()
#new_df['Cleaned Company Name'].value_counts()
#new_df['Location'].value_counts()
#new_df['Size'].value_counts()
new_df['Easy Apply'].value_counts()
```

```python
# Function to split the "Location" column into "City" and "State"
def split_location(location):
    parts = location.split(', ')
    if len(parts) == 2:
        return parts[0], parts[1]
    else:
        return None, None

# Apply the function to the "Location" column and create new columns
new_df['City'], new_df['State'] = zip(*new_df['Location'].apply(split_location))

# Drop the original "Location" column
new_df = new_df.drop(columns=['Location'])
```

```python
#Grab data analyst roles from Texas only
texas = new_df.loc[new_df['State'] == 'TX']
```
